# Statistical Tests

```python
# Importing Statistical Tools
import scipy.stats as stats
from statsmodels.formula.api import ols
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
```

```python
# FEV1 and FVC  are important in Asthma diagnosis and often used together as a ratio
# Asssessing their statistical relationship might give us more insight
```

```python
#Correlation Test (Pearson and Spearman)
#h0(Pearson) there is a no correlation between the variables
#h0(Spearman) there is a no correlation between the variable
pearson_corr = stats.pearsonr(data['LungFunctionFEV1'], data['LungFunctionFVC'])
spearman_corr = stats.spearmanr(data['LungFunctionFEV1'], data['LungFunctionFVC'])
print("Pearson Correlation:", pearson_corr)
print("Spearman Correlation:", spearman_corr)
```

```python
# since there is no correlation between the 2 varibles we have to use them both as independent features
# instead of the FEV1/FVC ratio normally used to diagnose Asthma
```

# Feature Engineering

```python
# re-exploring the target feature, checking the distribution of diagnosis
diagnosis_distribution = data['Diagnosis'].value_counts()
print(diagnosis_distribution)
```

```python
# plotting the distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Diagnosis', data=data)
plt.title('Distribution of Diagnosis')
plt.show()
```

```python
# positive diagnosis is less than 5% of the data
# therefore, resampling using the SMOTE technique must be done to balance the data
from imblearn.over_sampling import SMOTE
```

```python
# perform SMOTE resampling
feature = ['Age', 'Gender', 'Ethnicity', 'BMI', 'Smoking', 'PhysicalActivity',
       'DietQuality', 'SleepQuality', 'PollutionExposure', 'PollenExposure',
       'DustExposure', 'PetAllergy', 'FamilyHistoryAsthma',
       'HistoryOfAllergies', 'Eczema', 'HayFever', 'GastroesophagealReflux',
       'LungFunctionFEV1', 'LungFunctionFVC', 'Wheezing', 'ShortnessOfBreath',
       'ChestTightness', 'Coughing', 'NighttimeSymptoms', 'ExerciseInduced']
x = data[feature]
y = data['Diagnosis']

# Apply SMOTE to balance the dataset
smote = SMOTE()
x_resampled, y_resampled = smote.fit_resample(x, y)

# Check the distribution after applying SMOTE
resampled_distribution = pd.Series(y_resampled).value_counts()
print(resampled_distribution)

# Plot the distribution after SMOTE
plt.figure(figsize=(10, 6))
sns.countplot(x=y_resampled)
plt.title('Distribution of Alert_11 after SMOTE')
plt.show()
```

### Splitting data into train and test data

```python
# import libraries for train test splitting and randomforrest classifier
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_resampled,y_resampled,test_size=0.2,train_size=0.8,random_state=42)
```

```python
### Build and Fit the Random Forest Model
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
rf_pred=rf.predict(x_test)
```

```python
rf_pred
```

# ROC curve to compare all three models

```python
# import the ROC, AUC library
from sklearn.metrics import roc_curve, auc
```

```python
# Predicting probabilities
rf_pred = rf.predict_proba(x_test)[:, 1]
tr_pred = tr.predict_proba(x_test)[:, 1]
sv_pred = sv.predict_proba(x_test)[:, 1]
```

```python
# Calculating ROC curve
rf_fpr, rf_tpr, _ = roc_curve(y_test, rf_pred)
tr_fpr, tr_tpr, _ = roc_curve(y_test, tr_pred)
sv_fpr, sv_tpr, _ = roc_curve(y_test, sv_pred)
```

```python
# Calculating AUC
rf_auc = auc(rf_fpr, rf_tpr)
tr_auc = auc(tr_fpr, tr_tpr)
sv_auc = auc(sv_fpr, sv_tpr)
```

```python
# Plotting the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(rf_fpr, rf_tpr, label=f'RandomForestClassifier (AUC = {rf_auc:.3f})')
plt.plot(tr_fpr, tr_tpr, label=f'DecisionTreeClassifier (AUC = {tr_auc:.3f})')
plt.plot(sv_fpr, sv_tpr, label=f'SupportVectorClassifier (AUC = {sv_auc:.3f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```
